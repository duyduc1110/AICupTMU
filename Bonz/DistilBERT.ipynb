```python
import transformers
import torch

from transformers import BertTokenizer, BertModel, BertConfig, DistilBertForSequenceClassification, BertForPreTraining
```

```python
config = BertConfig(output_hidden_states=True)
t = BertTokenizer.from_pretrained('bert-base-uncased')
    model = BertModel(config)

input_ids = torch.tensor(t.encode("Hello, my dog is cute ha ha")).unsqueeze(0)
outputs = model(input_ids)
last_hidden_states = outputs[0]
```
